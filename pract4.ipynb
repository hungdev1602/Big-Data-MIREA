```python
import math

ZS = 10
dZ = 10
N = 50
dT = 0.01

W = int(ZS / dZ)

dN = 1 / N 
N_list = []

def D(mul, k):
    return 1 - (math.e ** -(mul * k))

def d_round(num, k):
    return round(D(N_list[num - 1], k), 3)

# Заполняем список N_list
for i in range(N):
    N_list.append(i * dN + 0.1)

k1 = k2 = k3 = k4 = 1 
d1 = d2 = d3 = d4 = 0 
op_D4 = []
op_K4 = [] 
count = 0
max_D = 0

for k1 in range(W + 1): 
    for k2 in range(W + 1):
        for k3 in range(W + 1):
            for k4 in range(W + 1):
                if ((k1 + k2 + k3 + k4) == W):
                    count += 1
                    d1 = d_round(1, k1) 
                    d2 = d_round(2, k2) 
                    d3 = d_round(3, k3) 
                    d4 = d_round(4, k4)
                    if d1 + d2 + d3 + d4 > max_D: 
                        max_D = d1 + d2 + d3 + d4 
                        op_D4.clear() 
                        op_D4.append(d1) 
                        op_D4.append(d2) 
                        op_D4.append(d3) 
                        op_D4.append(d4)

                        op_K4.clear() 
                        op_K4.append(k1) 
                        op_K4.append(k2) 
                        op_K4.append(k3) 
                        op_K4.append(k4)

print("Количество комбинаций:", count) 
print("Время расчёта:", count * dT, "сек") 
print("Max D:", round(max_D, 3)) 
print("Оптимальные значения D:", op_D4) 
print("Оптимальные затраты K:", op_K4)
```
