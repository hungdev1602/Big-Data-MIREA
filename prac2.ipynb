```python
import numpy as np
import matplotlib.pyplot as plt

#    1
S1, S2 = 1.0, 0.8
N0, Nk, dN = 110, 5000, 50
N_values = np.arange(N0, Nk + dN, dN)

#   -
def gustafson_barsis_law(N, S):
  return N + (1 - N) * S
#    S1  S2
speedup_S1 = gustafson_barsis_law(N_values, S1)
speedup_S2 = gustafson_barsis_law(N_values, S2)
#  
plt.figure(figsize=(10, 6))
plt.plot(N_values, speedup_S1, label=f'S = {S1}', linewidth=2)
plt.plot(N_values, speedup_S2, label=f'S = {S2}', linewidth=2)
plt.xlabel('  (N)')
plt.ylabel('')
plt.title(' 1:    N   S\n( -)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
